## 6.Categorical Column Exploration

```python
# Machine Type Distribution
type_counts = df['Type'].value_counts().sort_index()
type_pct = df['Type'].value_counts(normalize=True).sort_index()*100

type_summery = pd.DataFrame({
    'Machine Type': type_counts.index,
    'Count': type_counts.values,
    'Percentage (%)': type_pct.values.round(2)
})

print(type_summery.to_string(index=False))
print(f'\nTotal unique machine types: {df["Type"].nunique()}')
```

## 7.Target Variable Analysis - Machine Failure

```python
# Machine Falure Distribution
failure_counts = df['Machine failure'].value_counts().sort_index()
failure_pct = df['Machine failure'].value_counts(normalize=True).sort_index() * 100

print("Machine Failure Distribution: ")
failure_summery = pd.DataFrame({
    'Status': ['No Failure (0)','Failure (1)'],
    'Count': failure_counts.values,
    'Percentage (%)': failure_pct.values.round(2)
})

print(failure_summery.to_string(index=False))
```

**Imbalance Data Detected** - Failures are only 3.39% of total data.

## 8.Failure Type Breakdown

```python
# Failure Type Defination
failure_type_names = {
    'TWF': 'Tool Wear Failure',
    'HDF': 'Heat Dissipation Failure',
    'PWF': 'Power Failure',
    'OSF': 'Overstrain Failure',
    'RNF': 'Random Failure'
}

failure_types = ['TWF', 'HDF', 'PWF', 'OSF', 'RNF']
total_failures = df['Machine failure'].sum()

print("Failure Type Breakdown\n")

failure_df = pd.DataFrame({
    'Code': failure_types,
    'Failure Type': [failure_type_names[f] for f in failure_types],
    'Count': [df[f].sum() for f in failure_types],
    'Percentage of Total Failures (%)': [(df[f].sum() / total_failures * 100).round(2) for f in failure_types]
})

failure_df = failure_df.sort_values('Count', ascending=False).reset_index(drop=True)
print(failure_df.to_string(index=False))
```

```python
print(f'Total Failures: {int(total_failures)}')
print(f'Total Faiure Type Entries: {df[failure_types].sum().sum()}')
```

```python
# check failure without specific types
no_type_failures = df[(df['Machine failure'] == 1) & (df['Machine failure'] == 0) ]

print("Failure without specific type")
print(f'Count {len(no_type_failures)}')

if len(no_type_failures) > 0:
    print(f'\n These are failures where no specific type is recorded.')
else:
    print('\n All failures have a specific failure type assigned.')
```
